```python
data = []
k = 0
#변수를 초기화 한다

#무한루프로 사용자가 원하는 만큼 숫자를 받아 리스트에 추가한다.
while 1 :
    print(k + 1, '번째')
    temp = input('data 입력 \'끝\'입력시 종료 : ') #입력을 우선 문자열로 저장한다
    
    #무한루프에서 사용자가 "끝" 을 입력하면 루프에서 빠져나온다.
    if temp == '끝' :
        
        break
        
        
    data.append(int(temp)) #temp 문자열을 int로 바꾼 후 data 리스트에 추가한다.
    
    k = k + 1 #문자열의 길이를 알기 위한 변수이다.
    
    
FLAG = 1  #순서 변경 체크를 위한 변수이다. 1일 경우 변경이 이루어진 것 0일 경우 변경이 없는 경우이다.

print('입력받은값 = ', data) #받은 데이터를 확인하기 위한 프린트 문이다.


#버블 소트 하는 구간이다.
while FLAG > 0 :
    
    k = k-1
    
    FLAG = 0 # while루프가 다시 돌 때마다 FLAG값을 0으로 초기화한다.
    
    for i in range(0, k, 1) :
        
        if data[i] > data[i + 1] : # 앞쪽 리스트 값이 뒤 리스트 값보다 크거나 같으면 if 문 안으로 들어온다.
            # 리스트 중 두 값을 비교 위치를 변경을 위해 앞쪽 리스트를 temp에 저장 후 뒤 리스트에 저장하고 다시 temp 값을 다시 뒤 리스트에 저장하는 식
            temp = data[i]
            data[i] = data[i + 1]
            data[i + 1] = temp
            ## data[i], data[i + 1] = data[i + 1], data[i]
            FLAG = 1 # if문 안으로 들어왔으니 변경이 이루어진 것으로 생각하고 FLAG 값을 1로 변경한다.
        print('Sort 과정 = ', data)
    print()
            
            
print('Sort 결과 = ', data) #결과를 출력하는 프린트 문이다.
```
