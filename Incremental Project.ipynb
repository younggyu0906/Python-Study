```python
import urllib.request
import string

urls = urllib.request.urlopen("http://cse.kut.ac.kr").read().decode("utf-8")
s = str(urls)
abstr = ''
bracketCheck = 0
sCharCheck = 0
for i in s :
    if bracketCheck == 0 :
        if i == '<' :
            bracketCheck = 1
        else :
            for j in string.punctuation :
                if i == j :
                    sCharCheck = 1
            if sCharCheck :
                sCharCheck = 0
            else :
                abstr += i
    if bracketCheck == 1 :
        if i == '>' :
            bracketCheck = 0

temp = abstr.split()
urldic = {temp[0]:1}
for i in temp :
    if i in urldic :
        urldic[i] += 1
    else :
        urldic[i] = 1

print(urldic)
```
